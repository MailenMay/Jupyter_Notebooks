```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
```

```python
# Desviación Estandar:
df['price_usd'].std()
```

```python
# Rango = valor max - valor min
rango = df['price_usd'].max() - df['price_usd'].min()
rango
```

```python
# Quartiles
median = df['price_usd'].median()
Q1 = df['price_usd'].quantile(q = 0.25)
Q3 = df['price_usd'].quantile(q = 0.75)
min_value = df['price_usd'].quantile(q = 0)
max_value = df['price_usd'].quantile(q = 1.0)
print(min_value, Q1, median, Q3, max_value)
```

```python
#Rango intercuartil iqr
iqr = Q3 - Q1
iqr
```


#### Limites para detección de outliers (datos simétricamente distribuidos)

Datos entre $Q_1- 1.5  \times  IQR$ y $Q_3 + 1.5 \times IQR$

```python
minlimit = Q1 - 1.5*iqr
maxlimit = Q1 + 1.5*iqr
print(minlimit,maxlimit)
```
