```python
import matplotlib.pyplot as plt #Libreria para graficar
import numpy as np #Libreria para manejo de vectores y utilidades matematicas.
```

### Funci贸n lineal

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$


```python
N = 100

m = -1

b = 3

def f(x):
    return m*x + b

x = np.linspace(-10,10,num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.grid() #Agrego cuadricula

plt.show()
```

### Funciones polin贸micas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una funci贸n que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polin贸micas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 

```python
def f(x):
    return x**7

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.grid() #Agrego cuadricula

plt.show()
```

### Funci贸n logaritmo

El logaritmo est谩 definido por la **relaci贸n**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que est谩 elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$





```python
def f(x):
    return np.log2(x)

x = np.linspace(0.001,256,num=1000)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.grid() #Agrego cuadricula

plt.show()
```

# Funci贸n seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funci贸n escal贸n de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
    Y = np.zeros(len(x))
    for idx,x in enumerate(x):
        if x >= 0:
            Y[idx]=1
    return Y

N = 1000
x = np.linspace(-10,10,num=N)
y = H(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.grid() #Agrego cuadricula

plt.show()
```

### Reto de Funcion Escalonada

$$F(x) = 
     \begin{cases}
        x^2+2x, &\quad \text{para } x \leq -1 \\
        x,  &\quad\text{para } -1<x\leq 1 \\
        -1, &\quad \text{para } x > 1 \\
     \end{cases}
$$

```python
def f(x):
    Y = np.zeros(len(x))
    for idx,x in enumerate(x):
        if x <= -1:
            Y[idx]=(pow(x,2)+2*x)
        elif (-1<x and x<=1):
            Y[idx] = x
        else:
            Y[idx] = -1

    return Y

N = 10
x1 = np.linspace(-10,-1,num=N)
x2 = np.linspace(-1,-1,num=N)
x3 = np.linspace(1,10,num=N)
#y = f(x)

fig, ax = plt.subplots()
#ax.plot(x,y)
plt.plot(x1, f(x1), 'g',  label='y = 2, x en [-2, 0]')
plt.plot(x2, f(x2), 'b', label='y = x + 2, x en [0, 1]')
plt.plot(1, -1, 'ro', markersize=5, fillstyle='none', markerfacecolor='w')
plt.plot(x3, f(x3), 'r', label='y = .5*x - 1.5, x en (1, 3]')
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.grid() #Agrego cuadricula

plt.show()
```
