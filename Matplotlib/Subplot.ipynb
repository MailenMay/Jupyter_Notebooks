```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#Creo una matriz de grafico de 1 fila y 2 columnas
plt.subplot(1,2,1) #Los parametros de Subplot son (Filas, Columnas,Index)
plt.plot(x,y,'r--')
plt.subplot(1,2,2) #Los parametros de Subplot son (Filas, Columnas,Index)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1) #Los parametros de Subplot son (Filas, Columnas,Index)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2) #Los parametros de Subplot son (Filas, Columnas,Index)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1) #Los parametros de Subplot son (Filas, Columnas,Index) Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b--')#Plot 2

plt.subplot(2,1,2) #Los parametros de Subplot son (Filas, Columnas,Index) Grafico 2
plt.pie(y) #Plot 1
plt.show()
```
