```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x) #Aplico la funcion seno a X
```

```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y,label="$Sin(x)$")
axes[0].set_title("Relacion X - Y")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend()# Con esto veo la leyenda

axes[1].plot(y,x)
axes[1].set_title("Relacion Y - X")
axes[1].set_xlabel("Y")
axes[1].set_ylabel("X")
```

```python
plt.plot(x,y,label="Sin(x)")
plt.title("Este es un titulo")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() #Uso loc = para indicar donde quiero dejar la leyenda. Si no indico nada, el lo coloca donde lo crea mas conveniente. Con bbox_to_anchor(x,y) indico donde lo quiero poner
```
