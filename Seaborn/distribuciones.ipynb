```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')#Dataset de seaborn

tips.head()
```

```python
sns.histplot(data = tips, x = 'tip', bins=15,cumulative= False,hue='sex',stat="count",multiple='dodge')#Se acumulen las distribuciones
#Stat como muestro Eje y, multiple como presento los datos
plt.show()
```

```python
#Grafico de densidad
sns.kdeplot(data = tips,x = 'tip',hue= 'sex',cumulative = False,shade = True,bw_adjust=1)#Cumulative = acumulacion//Shade= crea area debajo de la curva
plt.show()
```

```python
sns.ecdfplot(data = tips,x = 'tip',hue= 'sex',stat='count')
plt.show()
```

```python
#Puedo cambiar el tipo de grafica con kind
sns.displot(data = tips,x = 'tip',hue= 'sex',kind='hist',multiple= 'stack')
plt.show()
```

## Tip for Smoke

```python
#Puedo cambiar el tipo de grafica con kind
sns.displot(data = tips,x = 'tip',hue= 'smoker',kind = 'kde',fill=True)
plt.show()
```

## Tip of Day

```python
#Puedo cambiar el tipo de grafica con kind
sns.displot(data = tips, x = 'tip',hue= 'day',kind = 'kde',multiple='stack')
plt.show()
```
